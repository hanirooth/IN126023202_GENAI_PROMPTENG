```python

# Install basic package (optional)
!pip install -q pandas

import re
import pandas as pd

# ==========================================================
# JOB DESCRIPTION
# ==========================================================

job_description = """
Data Scientist Role

Required Skills:
Python
SQL
Machine Learning
Deep Learning
TensorFlow
PyTorch
Power BI
Pandas
AWS
3+ years experience
"""

# ==========================================================
# SAMPLE RESUMES
# ==========================================================

resumes = {
    "Strong Candidate": """
Rahul Sharma
4 years experience as Data Scientist

Skills:
Python, SQL, Machine Learning, Deep Learning, TensorFlow, AWS

Tools:
Power BI, Pandas, Jupyter
""",

    "Average Candidate": """
Amit Kumar
2 years experience as Analyst

Skills:
Python, SQL, Pandas

Tools:
Excel, Tableau
""",

    "Weak Candidate": """
Ravi Singh
Fresher

Skills:
Communication, MS Office

Tools:
Word, Excel
"""
}

# ==========================================================
# REQUIRED SKILLS
# ==========================================================

required_skills = [
    "python",
    "sql",
    "machine learning",
    "deep learning",
    "tensorflow",
    "pytorch",
    "power bi",
    "pandas",
    "aws"
]

# ==========================================================
# FUNCTIONS
# ==========================================================

def extract_experience(text):
    match = re.search(r'(\d+)\s*years', text.lower())
    if match:
        return int(match.group(1))
    return 0

def extract_skills(text):
    found = []
    lower_text = text.lower()

    for skill in required_skills:
        if skill in lower_text:
            found.append(skill.title())

    return found

def calculate_score(skills, exp):
    skill_score = (len(skills) / len(required_skills)) * 80
    exp_score = min(exp / 3, 1) * 20
    total = round(skill_score + exp_score)
    return min(total, 100)

def verdict(score):
    if score >= 80:
        return "Highly Recommended"
    elif score >= 60:
        return "Recommended"
    elif score >= 40:
        return "Consider"
    else:
        return "Rejected"

def explain(score, skills, exp):
    return f"""
Matched Skills: {len(skills)}
Experience: {exp} years
Reason: Candidate has relevant skills and experience.
"""

# ==========================================================
# RUN SCREENING
# ==========================================================

results = []

for name, resume in resumes.items():

    skills = extract_skills(resume)
    exp = extract_experience(resume)
    score = calculate_score(skills, exp)
    final_verdict = verdict(score)
    explanation = explain(score, skills, exp)

    print("="*60)
    print(name)
    print("="*60)

    print("Extracted Skills:", skills)
    print("Experience:", exp, "years")
    print("Fit Score:", score, "/100")
    print("Verdict:", final_verdict)
    print(explanation)

    results.append({
        "Candidate": name,
        "Score": score,
        "Verdict": final_verdict
    })

# ==========================================================
# SUMMARY TABLE
# ==========================================================

df = pd.DataFrame(results)

print("\nFINAL SUMMARY")
print(df)
```
